```python
from defs_graph_plot import (
    load_data, build_ranking, build_pairs,
    save_processed
)

# Laden
orders, data = load_data(
    "../data/orders.csv",
    "../data/order_products__prior.csv",
    "../data/products.csv"
)

ranking = build_ranking(data, orders)
pairs   = build_pairs(data)
save_processed(data, ranking, pairs)  # ← speichern
```

```python
import networkx as nx
from defs_graph_plot import (
    build_confidence_rules, build_lift_rules,
    build_lift_graph, build_plot, detect_communities,
    communities_to_dataframe, load_processed
)

# Ab hier immer starten:
data, ranking, pairs = load_processed()  # ← laden

# Regeln
confidence = build_confidence_rules(pairs, data, min_count=10)
lift       = build_lift_rules(pairs, data, min_count=100)

# Graph & Plot
G           = build_lift_graph(pairs, data, ranking, top_n=300, min_count=50, min_lift=2.0)
pos         = nx.spring_layout(G, dim=3, k=0.6, iterations=200, seed=42)
communities = detect_communities(G)
fig         = build_plot(G, pos, ranking, communities=communities)
fig.show()

communities_to_dataframe(communities)
```
